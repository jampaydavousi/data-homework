# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#Grouping school data by school name
group_school_data = school_data_complete.groupby(["school_name"])
group_school_data.count().head()
```

```python
# Dividing the toal budget per school by number of students in that school
budget_per_student_per_school = school_data["budget"]/school_data["size"]
budget_per_student_per_school
```

```python
# Using the mean function to calculate the average of the math score list
average_grouped_math = group_school_data["math_score"].mean()
average_grouped_math
```

```python
# Using the mean function to calculate the average of the reading score list
average_grouped_reading = group_school_data["reading_score"].mean()
average_grouped_reading
```

```python
# determining number of students who took the math exam
group_school_data["math_score"].count()
```

```python
# Determining the number of students per each school withe a passing math grade
grouped_passing_math = (school_data_complete["math_score"] >=70).groupby(school_data_complete["school_name"]).sum().astype(int)
grouped_passing_math
```

```python
#Dividing the list of students withe passing math grade by total number of students at each school.
percent_passing_math = (school_data_complete["math_score"] >=70).groupby(school_data_complete["school_name"]).sum()/group_school_data["math_score"].count()
percent_passing_math
```

```python
# determining number of students who took the reading exam
group_school_data["reading_score"].count()
```

```python
# Determining the number of students per each school withe a passing reading grade
grouped_passing_reading = (school_data_complete["reading_score"] >=70).groupby(school_data_complete["school_name"]).sum().astype(int)
grouped_passing_reading
```

```python
#Dividing the list of students withe passing reading grade by total number of students at each school.
percent_passing_reading = (school_data_complete["reading_score"] >=70).groupby(school_data_complete["school_name"]).sum()/group_school_data["reading_score"].count()
percent_passing_reading
```

```python
#calcualting the overall passing rate
overall_passing_rate = (percent_passing_math + percent_passing_reading)/2
overall_passing_rate
```

```python
# Creating a new series to inspect each data series that would ultimately end up in the completed data frame grouped by school
schoolsummary = [
                    ['School Name', group_school_data['school_name']],
                    ['School Type', group_school_data["type"]],
                    ['Total Students', group_school_data["size"]],
                    ['Total School Budget', group_school_data["budget"]],
                    ['Per Student Budget' , percent_passing_math],
                    ['Average Math Score' , group_school_data["math_score"].mean()],
                    ['Average Reading Score' , group_school_data["reading_score"].mean()],
                    ['% Passing Math' , percent_passing_math],
                    ['% Passing Reading' , percent_passing_reading],
                    ['Overall Passing Rate' , overall_passing_rate]
        ]
schoolsummary
```

```python
# Creating a new data frame ...ONE to be merged later
school_summary_one = pd.DataFrame(
                {
                    "School Name": school_data['school_name'],
                    "School Type": school_data["type"],
                    "Total Students": school_data["size"],
                    "Total School Budget": school_data["budget"],
                    "Per Student Budget" : budget_per_student_per_school
                    }
)
school_summary_one
```

```python
# Creating a new data frame ...TWO to be merged later
school_summary_two = pd.DataFrame(
                {
                    "Average Math Score": average_grouped_math,
                    "Average Reading Score": average_grouped_reading,
                    "% Passing Math": percent_passing_math,
                    "% Passing Reading": percent_passing_reading,
                    "Overall Passing Rate": overall_passing_rate
                }
)
school_summary_two
```

```python
#renaming the index name 'school-name' to "School Name" so School Name can be used as the common column for merging
school_summary_two.index.name = "School Name" #rename(columns = {'School Name' : 'school_name'}, inplace=True)
school_summary_two
```

```python
#merging summary one data frame with summary two data frame so they can be merged
school_summary_complete = pd.merge(school_summary_one, school_summary_two, how="left", on=["School Name", "School Name"])
school_summary_complete
```

```python
#formatting the columns with %
school_summary_complete.style.format({
                                        'Total Student Budget': '$'.format,
                                        'Average Math Score': '{:,.2f}'.format,
                                        'Average Reading Score': '{:,.2f}'.format,
                                        '% Passing Math': '{:,.3%}'.format,
                                        '% Passing Reading': '{:,.3%}'.format,
                                        'Overall Passing Rate': '{:,.3%}'.format
}
)
```

```python
#exporting this data frame to excel
school_summary_complete.to_excel("SchoolSummarybySchools.xlsx")
```

## Scores by School Spending

  * Average Reading Score

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
# recalling the header of the complete grouped school data frame 
school_summary_complete.head()
```

```python
# assign a new column heading for the bins
school_data_complete_plus["Spending Ranges (Per Student)"] = pd.cut(school_data_complete_plus["Spending Per"], spending_bins, labels=group_names)
school_data_complete_plus.head()
```

```python
#Grouping school data by spending per student bins
group_school_data_plus = school_data_complete_plus.groupby(["Spending Ranges (Per Student)"])
group_school_data_plus.count().head()
```

```python
# calcualting average math score 
average_grouped_plus_math = group_school_data_plus["math_score"].mean()
average_grouped_plus_math
```

```python
# calcualting average reading score 
average_grouped_plus_reading = group_school_data_plus["reading_score"].mean()
average_grouped_plus_reading
```

```python
#Calculating the list of students with passing math grade
percent_passing_plus_math = (school_data_complete_plus["math_score"] >=70).groupby(school_data_complete_plus["Spending Ranges (Per Student)"]).sum()/group_school_data_plus["math_score"].count()
percent_passing_plus_math
```

```python
#Calculating the list of students with passing reading grade
percent_passing_plus_reading = (school_data_complete_plus["reading_score"] >=70).groupby(school_data_complete_plus["Spending Ranges (Per Student)"]).sum()/group_school_data_plus["reading_score"].count()
percent_passing_plus_reading
```

```python
#calcualting the overall passing rate
overall_passing_rate_plus = (percent_passing_plus_math + percent_passing_plus_reading)/2
overall_passing_rate_plus
```

```python
# Creating a new data frame
school_summary_spending = pd.DataFrame(
                {
                    "Average Math Score": average_grouped_plus_math,
                    "Average Reading Score": average_grouped_plus_reading,
                    "% Passing Math": percent_passing_plus_math,
                    "% Passing Reading": percent_passing_plus_reading,
                    "% Overall Passing Rate": overall_passing_rate_plus
                }
)
school_summary_spending
```

```python
# formating the columns with percentage
school_summary_spending['% Passing Math'] = school_summary_spending['% Passing Math']*100
school_summary_spending['% Passing Reading'] = school_summary_spending['% Passing Reading']*100
school_summary_spending['% Overall Passing Rate'] = school_summary_spending['% Overall Passing Rate']*100
school_summary_spending
```

```python
#exporting this data frame to excel
school_summary_spending.to_excel("SchoolSummarybySpending.xlsx")
```
